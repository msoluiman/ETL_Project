```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

state_crime = "Resources/state_crime.csv"
state_crime_results = pd.read_csv(state_crime)

state_crime_results
```

```python
state_crime_results.info()
```

```python
state_crime_tab = state_crime_results.loc[state_crime_results["year"] == 2019,["state", "year", "region", "violent_crime_in_state", "embezzlement_in_state"]]
state_crime_tab.head()
```

```python
state_crime_tab.dtypes
```

```python
state_edu = "Resources/state_education.csv"
state_edu_results = pd.read_csv(state_edu)

state_edu_results
```

```python
state_edu_results.dtypes
```

```python
state_edu_tab = state_edu_results[["state", "state_median_income", "state_proportion_with_bachelors_or_higher"]]
state_edu_tab.head()
```

```python
combined_data_df = pd.merge(state_crime_tab, state_edu_tab, how="left", on=["state"])
combined_data_df.head()
```

```python
combined_data_df.rename(columns={"violent_crime_in_state":"Violent_Crime", "embezzlement_in_state":"Embezzlement", 
                                 "state_median_income":"Median_Income", "state_proportion_with_bachelors_or_higher":"Bachelors_Higher"})
```
